```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Score =[5,4,3,2,1,-1,-2,-3,-4,-5]
City = ('Amsterdam','Sydney','Vienna','Stockholm','Copenhagen','Milan','Washington DC','New York','Moscow','Mexico city')
x_pos = np.arange(len(City))
plt.bar(x_pos, Score, color=['red','yellow','pink','green','purple','orange','blue','grey','brown','black'])
plt.title('Top 5 and Bottom 5 Healthy Cities 2021', color='darkblue', fontsize=16)
plt.xlabel('CITY', color='darkblue', fontsize=16)
plt.ylabel('SCORE', color='darkblue',fontsize=16)
plt.xticks(x_pos, City, color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
x=[1434,1712,1501,1452,1380,1718,1779,1779,1965,2137]
y=[5,4,3,2,1,-1,-2,-3,-4,-5]
annotations=["Amstdam", "Sydney", "Vienna", "Stockh", "Copenh", "Milan", "WashDC", "NY", "Moscow", "MexicoC"]
plt.title("Correlation of Annual Average Hours Worked variable to Score", color="#9A0EEA", fontsize=16)
plt.scatter(x,y)
plt.plot(x,y, marker="v", color="#C20078")
plt.xlabel('ANNUAL AVERAGE HOURS WORKED', color="#7E1E9C", fontsize=14)
plt.ylabel('SCORE', color="#7E1E9C", fontsize=14)
for i, label in enumerate(annotations):
    plt.annotate(label, (x[i], y[i]), color="#FF81C0")
plt.show()
```
